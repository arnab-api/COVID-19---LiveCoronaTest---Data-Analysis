```python
import json
import pandas as pd
```

```python
with open("dist_matrix_unformatted", "r") as f:
    lines = f.readlines()
    district_names = []
    dist_arr_2d = {}
    for line in lines:
        source_name = " ".join(line.split(" ")[0:-64])
        dist_arr = line.split(" ")[-64:]
        # print(source_name, len(dist_arr))
        district_names.append(source_name)
        dist_arr_2d[source_name] = dist_arr

    dist_matrix = {}
    for source in dist_arr_2d:
        dist_arr = dist_arr_2d[source]
        dist_matrix[source] = {}
        for dst in range(len(district_names)):
            destination = district_names[dst]
            dist_matrix[source][destination] = int(dist_arr[dst].strip())

dist_matrix

```

```python
df = pd.DataFrame(dist_matrix)
df
```

```python
df.to_csv("Bangladesh district-wise distance matrix.csv")
```

```python
with open("Bangladesh district-wise distance matrix.json", "w") as f:
    json.dump(dist_matrix, f)
```
