```python
cnt=0

for sn in symptom_list:
    
#     if cnt>1:
#         break
        
    df = pd.read_csv('/mnt/mamin17/bsbijoy2050/Covid19/bdcovid/new_data_11_march/Shajib-Sir-Formatted-Data/'+sn+'_Reports_over_Districts.csv', header  = 0)
    print(sn, df.shape)
    
    
    newcol = []
    for ind in range(df.shape[0]):
        newcol.append(0)
    
    for sin_dist in dist_list:
        if sin_dist in df.columns.tolist():
            continue
        else:
            df[sin_dist] = newcol
    df = df[dist_list]
    
#     if sn == 'Cancer':
        
#         line = pd.DataFrame(newrow, index=[179])
#         df = pd.concat([df.loc[:178], line, df.loc[179:]]).reset_index(drop=True)
    
    print(sn, df.shape)
    
    for singleDate in date_list:
        if singleDate in list(df['date']):
            continue
        else:
            newrow = {}
            
            for sin_dist in dist_list[:-1]:
                newrow[sin_dist] = 0
                
            newrow['date'] = singleDate
            
            line = pd.DataFrame([newrow])
            df = df.append(line, ignore_index=True)
#             df = pd.concat([line]).reset_index(drop=True)
        # newrow
    print(sn, df.shape)
    
    
    
    df['date'] =  pd.to_datetime(df['date'],format='%Y-%m-%d') 
    df = df.sort_values(by = ['date'], ascending = [True])
    display(df)
    df.to_csv('/mnt/mamin17/bsbijoy2050/Covid19/bdcovid/new_data_11_march/Processed/'+sn+'.csv', index = False)
    
    
#     dist_set.update(df.columns)
# #     df['date'] =  pd.to_datetime(df['date'],format='%Y-%m-%d') 
#     df = df[fil_col]
#     df.to_csv('/mnt/mamin17/bsbijoy2050/Covid19/bdcovid/'+sn+'.csv', index = False)
#     display(df)
    cnt+=1
```
