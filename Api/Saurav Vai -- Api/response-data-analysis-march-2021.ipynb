```python
limit = -1

sum_score = 0
count = 0

age_range=['0-20', '21-30', '31-40', '41-50', '51-65', '65+']
age_sum_risk = {}
age_sum_count = {}

for age in age_range:
    age_sum_risk[age] = 0
    age_sum_count[age] = 0

for index, row in df[['score_user_qa_', 'risk', 'answer_en']].sample(frac=1).iterrows():
    # print(row)
    sum_score += row['score_user_qa_']
    count += 1

    age = row['answer_en']

    # print(count, limit, " >> ", age)

    if(age not in age_range):
        continue
    
    age_sum_risk[age] += row['score_user_qa_']
    age_sum_count[age] += 1

    limit -= 1
    if(limit == 0):
        break

    if(count % 100000 == 0):
        print("completed {} of {}".format(count, df.shape))

print("Count: ", count)
print("Limit: ", limit)
print("average score: {}".format(round(sum_score/count, 2)))
for age in age_range:
    print('{} >> Total: {}, Count: {} >> Average: {}'.format(age, age_sum_risk[age], age_sum_count[age], age_sum_risk[age]/age_sum_count[age]))
```

```python
def plot_symptoms_over_risks(question_id, merged_df, symptoms_title, full_data=False, change_text_to_id = False):
    risks = ['low', 'mid', 'high']
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
    fig.suptitle(symptoms_title)
    axis = [ax1, ax2, ax3]
    for risk, ax in zip(risks, axis):
        
        if question_id == 3:
            merged_df = merged_df.loc[merged_df['option_id']] == 5
        this_risk_df = merged_df.loc[merged_df['risk'] == risk]

        this_risk_df_for_q = this_risk_df.loc[this_risk_df['question_id_user_qa_'] == question_id]
        if change_text_to_id:
            group_by_res = this_risk_df_for_q.groupby(['option_id']).size()
        else:
            group_by_res = this_risk_df_for_q.groupby(['answer_en']).size()
#         print(group_by_res)
        if change_text_to_id:
            group_by_res.plot.pie(y='mass', ax=ax,figsize=(12, 6), fontsize=12, autopct='%1.1f%%', title=risk)
        else:
            group_by_res.plot.pie(y='mass', ax=ax,figsize=(12, 6), fontsize=12, autopct='%1.1f%%', title=risk, labels=None, legend=True)
        
        ax.set_ylabel('')
    
    if change_text_to_id:
        answers = answer_df.loc[answer_df['question_id'] == question_id]
#         print(answers[['id','answer_bn']])
#         HTML(answers[['id','answer_bn']].to_html(index=False))
    plt.figure(figsize = (20,15))
    plt.savefig('{}/{}.png'.format(plot_dir, question_id))
    plt.show()
    
    
    df = merged_df.loc[merged_df['question_id_user_qa_'] == question_id]
    df = df.groupby(['risk','option_id' , 'answer_en']).size().reset_index()
    df.rename(columns = {list(df.keys())[-1]:'count'}, inplace = True)
    
    

    display(HTML(df.to_html()))
```

```python
def plot_symptoms_over_risks_on_full_data(question_id, merged_df, symptoms_title, change_text_to_id = False):
    fig, ax = plt.subplots(1, 1)
#     fig.suptitle(symptoms_title)
    this_risk_df_for_q = merged_df.loc[merged_df['question_id_user_qa_'] == question_id]
    group_by_res = this_risk_df_for_q.groupby(['answer_en']).size()
    if change_text_to_id:
        group_by_res.plot.pie(y='mass', ax=ax,figsize=(12, 6), fontsize=12, autopct='%1.1f%%',startangle=90, title=symptoms_title)
    else:
        group_by_res.plot.pie(y='mass', ax=ax,figsize=(12, 6), fontsize=12, autopct='%1.1f%%',startangle=90, title=symptoms_title, labels=None, legend=True)

#     group_by_res.plot.pie( ax=ax,figsize=(12, 6), fontsize=12, autopct='%1.1f%%', title=symptoms_title)
    ax.set_ylabel('')
    if change_text_to_id:
        answers = answer_df.loc[answer_df['question_id'] == question_id]
#         print(answers[['id','answer_bn']])
#         HTML(answers[['id','answer_bn']].to_html(index=False))
    plt.savefig('{}/{}_full.png'.format(plot_dir,question_id))
    plt.show()
    
    
    df = merged_df.loc[merged_df['question_id_user_qa_'] == question_id]
    df = df.groupby(['risk','option_id' , 'answer_en']).size().reset_index()
    df.rename(columns = {list(df.keys())[-1]:'count'}, inplace = True)
    
    

    display(HTML(df.to_html()))
```
