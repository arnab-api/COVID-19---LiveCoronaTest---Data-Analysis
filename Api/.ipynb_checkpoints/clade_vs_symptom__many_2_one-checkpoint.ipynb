# Read Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
clade_df = pd.read_csv("../June_mutation_freq.csv")
clade_df.head()
```

```python
clade_df.shape
```

```python
mutation_arr = set(list(clade_df['mutation']))
len(mutation_arr)
```

```python
clade_info = {}

for index, row in clade_df.iterrows():
    mutation_name = row['mutation'].split(":")[0]
#     mutation_name = row['mutation']
    month = row['month']
    if(mutation_name not in clade_info):
        clade_info[mutation_name] = {}
    if(month not in clade_info[mutation_name]):
        clade_info[mutation_name][month] = 0
    clade_info[mutation_name][month] += row["frequency"]
```

```python
clade_info
```

```python
month_arr = list(set(list(clade_df['month'])))
month_arr = sorted(month_arr)
```

```python
mutation_freq_arr = {}

for mutation in clade_info:
    mutation_freq_arr[mutation] = np.zeros(len(month_arr))
    for i in range(len(month_arr)):
        month = month_arr[i]
        if(month in clade_info[mutation]):
            mutation_freq_arr[mutation][i] = clade_info[mutation][month]
```

```python
print(month_arr)
mutation_freq_arr
```

```python
plt.rcParams["figure.figsize"] = (20,10)

# limit = 14
# plt.plot(month_arr[:limit], x, label="X")
# plt.plot(month_arr[:limit], y, label="Y")
# plt.legend(loc="upper right")

for mutation in mutation_freq_arr:
    plt.plot(month_arr, mutation_freq_arr[mutation], label=mutation)

plt.legend(loc="upper right")
plt.show()
```

```python
symptom_df = pd.read_csv("../symptom_frequency_monthWise - symptom_frequency_monthWise.csv")
symptom_df
```

```python
symptom_freq_arr = {}

for index, row in symptom_df.iterrows():
    symptom_name = row['Symptom Name']
    symptom_freq_arr[symptom_name] = np.zeros(len(month_arr))
    for i in range(len(month_arr)):
        month = month_arr[i]
        if(month in row):
            symptom_freq_arr[symptom_name][i] = row[month]
```

# Regression

```python
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
```

```python
def applyMinMaxNorm(arr):
    arr = np.array(arr)
    ret = np.zeros(len(arr))
    
    mx = arr.max()
    mn = arr.min()

    for i in range(len(arr)):
        ret[i] = (arr[i] - mn)/(mx - mn)
    
    return ret
```

```python
symptom = 'High fever'
mutation1 = 'ORF1b'
# mutation2 = 'ORF3a'

limit = 11

x1 = applyMinMaxNorm(mutation_freq_arr[mutation1][:limit])
x2 = applyMinMaxNorm(mutation_freq_arr[mutation2][:limit])
y = applyMinMaxNorm(symptom_freq_arr[symptom][:limit])
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(month_arr[:limit], x1, label="X1", color="blue")
plt.plot(month_arr[:limit], x2, label="X2", color="blue")
plt.plot(month_arr[:limit], y, label="Y", color="red")
plt.legend(loc="upper right")
```

# Multiple Regression

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
def getFeatureVector(feature_arr):
    ln = len(feature_arr[0])
    ret = []
    for i in range(ln):
        ret.append([])
        
    for arr in feature_arr:
        for i in range(len(arr)):
            ret[i].append(arr[i])
    return np.array(ret)
```

```python
print(getFeatureVector([x1,x2, x1]).shape)
getFeatureVector([x1,x2, x1])
```

```python
regr = linear_model.LinearRegression()
X = getFeatureVector([x1,x2])
regr.fit(X, y)
```

```python
print(regr.coef_)
print(regr.intercept_)
print(regr.score(X,y))
```

```python
y_pred = regr.predict(X)
mean_squared_error(y_pred, y)
```

# Symptom vs Clade Regressin Values

```python
from operator import itemgetter
```

```python
def getSymptomVsMutationReg(symptom, mutation, limit = 11):
    x = applyMinMaxNorm(mutation_freq_arr[mutation][:limit])
    y = applyMinMaxNorm(symptom_freq_arr[symptom][:limit])
    
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x,y)
    return {
        "slope": slope,
        "intercept": intercept,
        "r_squared": r_value**2,
        "p_value": p_value,
        "std_err": std_err
    }
```

```python
track_info = []
dct2D = {}

for symptom in symptom_freq_arr:
    for mutation in mutation_freq_arr:
        reg_info = getSymptomVsMutationReg(symptom, mutation)
        r_squared = reg_info['r_squared']
        print(symptom, mutation, " >>>>>>>> ", r_squared)
        track_info.append({
            "symptom": symptom,
            "mutation": mutation,
            "r_squared": r_squared
        })
        
        if(symptom not in dct2D):
            dct2D[symptom] = {}
        dct2D[symptom][mutation] = r_squared
```

```python
sorted(track_info, key=itemgetter('r_squared'), reverse=True)
```

```python
def plotMatrix(dct2D):
    # print(dct2D)
    arr2D = []
    row = []
    col = []
    for ik in dct2D:
        row.append(ik)
        arr1D = []
        for jk in dct2D[ik]:
            arr1D.append(round(dct2D[ik][jk], 3))
            if(jk not in col):
                col.append(jk)
        arr2D.append(arr1D)
    df_cm = pd.DataFrame(arr2D, index = row, columns = col)
    sns.heatmap(df_cm, annot=True)
    plt.show()
```

```python
plt.figure(figsize = (20,10))
plt.rcParams.update({'font.size': 18})
plotMatrix(dct2D)
```
