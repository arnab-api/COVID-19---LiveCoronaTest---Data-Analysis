```python
import pandas as pd
import os
import json
```

```python
path = "../New Data"
```

```python
os.listdir(path)
```

```python
df = pd.read_csv(path + "/High fever_Reports_over_Districts.csv")
print(df.shape)
df.head()
```

```python
district_arr = df.keys()[1:]
district_arr = list(district_arr)
```

```python
def getMonthTrackfromDF(df):
    month_track = {}

    for index,row in df.iterrows():
        date = row['date']
        month = date[:7]
        if(month not in month_track):
            month_track[month] = {
                'BD'   : 0,
                'Dhaka': 0
            }
        month_track[month]['Dhaka'] += row['Dhaka']
        summ = 0
        for district in district_arr:
            summ += row[district]
    #         print(district, row[district])
        month_track[month]['BD'] += summ
        
    return month_track
```

```python
getMonthTrackfromDF(df)
```

```python
new_data = {}

for csv in os.listdir(path):
    df = pd.read_csv(path + "/" + csv)
    district_arr = df.keys()[1:]
    district_arr = list(district_arr)
    print(district_arr)
    
    print(csv, " >>> ", df.shape)
    symptom = csv.split("_")[0]
#     print(symptom)
    new_data[symptom] = getMonthTrackfromDF(df)
```

```python
new_data
```

```python
with open("new_data.json", "w") as f:
    json.dump(new_data, f)
```

# Check Mutation Clade Association

```python
import os
import pandas as pd
```

```python
association_clade_mutation = pd.read_csv("../clade_mutation.csv", delimiter=";")
association_clade_mutation.head()
```

```python
keys = list(association_clade_mutation.keys())
keys
```

```python
import math
```

```python
clade_mutation = []
mutation_to_clade = {}
clade_to_mutation = {}

for index, row in association_clade_mutation.iterrows():
    obj = {}
    for key in keys:
        val = row[key]
        if(isinstance(val, float) and math.isnan(val)):
            continue
        if(key == 'aaSubstitutions'):
            val = val.split(",")
        print(key, " >>> ", val)
        obj[key] = val
    print()
    
    clade = obj['clade']
    if(clade not in clade_to_mutation):
        clade_to_mutation[clade] = []
        
    if("aaSubstitutions" in obj):
        for mutation in obj["aaSubstitutions"]:
            if(mutation not in clade_to_mutation[clade]):
                clade_to_mutation[clade].append(mutation)

            if(mutation not in mutation_to_clade):
                mutation_to_clade[mutation] = []
            if(clade not in mutation_to_clade[mutation]):
                mutation_to_clade[mutation].append(obj['clade'])
```
