```python
import pandas as pd
import os
import json
```

```python
path = "../New Data"
```

```python
os.listdir(path)
```

```python
df = pd.read_csv(path + "/High fever_Reports_over_Districts.csv")
print(df.shape)
df.head()
```

```python
district_arr = df.keys()[1:]
district_arr = list(district_arr)
```

```python
def getMonthTrackfromDF(df):
    month_track = {}

    for index,row in df.iterrows():
        date = row['date']
        month = date[:7]
        if(month not in month_track):
            month_track[month] = {
                'BD'   : 0,
                'Dhaka': 0
            }
        month_track[month]['Dhaka'] += row['Dhaka']
        summ = 0
        for district in district_arr:
            summ += row[district]
    #         print(district, row[district])
        month_track[month]['BD'] += summ
        
    return month_track
```

```python
getMonthTrackfromDF(df)
```

```python
new_data = {}

for csv in os.listdir(path):
    df = pd.read_csv(path + "/" + csv)
    district_arr = df.keys()[1:]
    district_arr = list(district_arr)
    print(district_arr)
    
    print(csv, " >>> ", df.shape)
    symptom = csv.split("_")[0]
#     print(symptom)
    new_data[symptom] = getMonthTrackfromDF(df)
```

```python
with open("new_data.json", "w") as f:
    json.dump(new_data, f)
```
