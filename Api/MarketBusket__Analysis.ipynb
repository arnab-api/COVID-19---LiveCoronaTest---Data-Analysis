# The Filter Function

```python
def getID_2_answerDictionary(df, idfield='id', answerfield='answerEn'):
    ret = {}
    for index, row in df.iterrows():
        ret[row[idfield]] = row[answerfield].strip()
    return ret
```

```python
def getUserAnswers__FilterByQuestion(question_id, multiple = False, limit = -1):
    question = question_df[question_df['id'] == question_id]
    q_dict = getID_2_answerDictionary(question, answerfield='questionEn')
    print("question >> ", q_dict)
    print()

    options = answer_df[answer_df['question_id'] == question.iloc[0]['id']]
    option_dict = getID_2_answerDictionary(options)
    print("Options")
    print(json.dumps(option_dict, indent=2))


    user_responses = qa_df[qa_df['question_id'] == question.iloc[0]['id']]
    print("user respose shape >> ", user_responses.shape)

    count = 0

    answer_dict = {}
    
    for index, row in user_responses.iterrows():
        try:
            user_id = row['user_id']
            if(multiple == True):
                if user_id not in answer_dict:
                    answer_dict[user_id] = []
            answer = option_dict[row['option_id']]
            if(multiple == True):
                answer_dict[user_id].append(answer)
            else:
                answer_dict[user_id] = answer                

            limit -= 1
            if(limit == 0):
                break

            count += 1
            if(count % 100000 == 0):
                print("processed {} of {}".format(count, user_responses.shape))
        except:
            continue
    
    print("successfully processed {} of {}".format(count, user_responses.shape))
    return answer_dict
```

# Analysis

```python
age_track = getUserAnswers__FilterByQuestion(1, multiple=False)
```

```python
symptom_track = getUserAnswers__FilterByQuestion(8, multiple=True)
```

```python
possible_ages = [
  "0-20",
  "21-50",
  "51-65",
  "65+",
  "21-30",
  "31-40",
  "41-50"
]

possible_symptoms = [
  "Respiratory illness",
  "Heart diseases",
  "High blood pressure",
  "Diabetes",
  "Kidney problem",
  "Cancer",
  "None of those"
]


association = {}
for age in possible_ages:
    association[age] = {}
    for symptom in possible_symptoms:
        association[age][symptom] = 0

user_count = 0
for user in symptom_track:
    symptom_arr = symptom_track[user]
    if(user not in age_track):
      print("could not find in the age_track >> ", user)
      continue
    age = age_track[user]
    for symptom in symptom_arr:
        association[age][symptom] += 1
    user_count += 1
    if(user_count % 100000 == 0):
        print("user count >> {}".format(user_count))
```

```python
print(user_count)
```

```python
def numPeople__inAgeRange(age):
    symptom_dict = association[age]
    ret = 0
    for symptom in symptom_dict:
        ret += association[age][symptom]
    return ret 

numPeople__inAgeRange('21-30')
```

```python
def numPeople__withSymptom(symptom):
    ret = 0
    for age in possible_ages:
        ret += association[age][symptom]
    return ret

numPeople__withSymptom("Diabetes")
```

```python
significant = []

for age in possible_ages:
    num_people_with_age = numPeople__inAgeRange(age)
    for symptom in possible_symptoms:
        num_people_with_symptom = numPeople__withSymptom(symptom)

        support = association[age][symptom]/user_count
        confidence = association[age][symptom]/num_people_with_age
        
        support_age = num_people_with_age/user_count
        support_symp = num_people_with_symptom/user_count
        # print(support_age, support_symp)
        if((support_age * support_symp) == 0):
            lift = 'N/A'
        else:
            lift = support/(support_age * support_symp)
        print("{} > {}: support = {}, confidence = {}, lift = {}".format(age, symptom, support, confidence, lift))
        
        if(lift != 'N/A' and lift > 1):
            significant.append({
                'X' : age,
                'Y' : symptom,
                'support': support,
                'confidence': confidence,
                'lift': lift
            })

significant
```
